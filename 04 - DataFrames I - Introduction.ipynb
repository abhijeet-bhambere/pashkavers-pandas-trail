## Differences between Shared Methods
- The `sum` method adds a **Series's** values.
- On a **DataFrame**, the `sum` method defaults to adding the values by traversing the index (row values).
- The `axis` parameter customizes the direction that we add across. Pass `"columns"` or `1` to add "across" the columns.

### We can set any one of the DataFrame columns as the index 

```python
revenue = pd.read_csv("revenue.csv",index_col="Date")
revenue
```

#### The `axis` parameter helps control the direction in which the `sum` method traverses -- either down a column (default) OR along a row

```python
# Adds all non-NaN values across each column ie sums along the index for each of the column
revenue.sum()
revenue.sum(axis="index")
revenue.sum(axis=0)
```

```python
# Setting axis=1 dictates the sum method to add across each index ie. across the lenght of each row across all columns
revenue.sum(axis=1)
revenue.sum(axis="columns")
```

```python
# We can sum this series to get a grand total
revenue.sum(axis="columns").sum()
```

#### The ability to change the direction in which a method operates unlocks newer perspective
In this example , keeping the default axis param (0 OR "index") we get total revenue for each of the cities, enabling a city-wise comparison of revenue.<br> 
Changing axis to 1 (or "columns") gives us a monthly sum of revenue, enabling a month-wise comparison of total revenue.<br>
Both perspectives are important  as per the context or business requirement.

## Select One Column from a DataFrame
- We can use attribute syntax (`df.column_name`) to select a column from a **DataFrame**. The syntax will not work if the column name has spaces.
- We can also use square bracket syntax (`df["column name"]`) which will work for any column name.
- Pandas extracts a column from a **DataFrame** as a **Series**.
- The **Series** is a view, so changes to the **Series** *will* affect the **DataFrame**.
- Pandas will display a warning if you mutate the **Series**. Use the `copy` method to create a duplicate.

```python
# This is one way of selecting a column & showing as a Series
nba.Name
```

```python
type(nba.Name)
```

```python
# This is another method of fetching a Series 
nba["Name"]
```

#### NOTE that both methods give a View of the DataFrame -- so modifying it would modify to the original DataFrame

```python
names = nba["Name"]
names.head()
```

```python
names.iloc[0] = "This New Name"
names.head()
```

##### But checking on original `nba` DataFrame we se that it also has been changed

```python
nba.head()
```

#### Avoid this by taking a `copy` of the required series so it is decoupled from the original DataFrame

```python
nba = pd.read_csv("nba.csv")
```

```python
names = nba["Name"].copy()
names.iloc[0] == "A New Name"
```

```python
names.head()
```

#### As seen the original df remains unaffected & changes only apply to the copy

```python
nba.head()
```

## Drop Rows with Missing Values
- Pandas uses a `NaN` designation for cells that have a missing value.
- The `dropna` method deletes rows with missing values. Its default behavior is to remove a row if it has *any* missing values.
- Pass the `how` parameter an argument of "all" to delete rows where all the values are `NaN`.
- The `subset` parameters customizes/limits the columns that pandas will use to drop rows with missing values.

#### One of THE MOST useful methods in pandas when handling missing or NaN values in a dataset.

```python
nba = pd.read_csv("nba.csv")
nba
```

#### Use of the `dropna` method depends on the use case -- either we might be better off removing a row **entirely** if even one of its column values are NaN /missing.
#### OR it could be the case that we might want to retain a row that has NaN but its other column values could be useful for other aspects of our analysis.

```python
# This removes rows with even a single NaN column value
nba.dropna()
```

```python
# Notice only the last row is removed -- one that had all column values missing. 
nba.dropna(how="all")
```

#### The `subset` parameter allows for a conditional dropping of rows. It tells the `dropna` func to look for NaN vlaues in a particular column(s) in a DataFrame AND only then drop the row. 

```python
# Drop entire row only if College column has NaN value
nba.dropna(subset=["College"])
```

```python
# Drop entire row only if EITHER College OR Salary columns have NaN values
nba.dropna(subset=["College","Salary"])
```

## Fill in Missing Values with the fillna Method
- The `fillna` method replaces missing `NaN` values with its argument.
- The `fillna` method is available on both **DataFrames** and **Series**.
- An extracted **Series** is a view on the original **DataFrame**, but the `fillna` method returns a copy.

For example, we might want to replace the NaN with 0.0 in the Salary column instead of deleting the row

## Sort a DataFrame with the sort_values Method I
- The `sort_values` method sorts a **DataFrame** by the values in one or more columns. The default sort is an ascending one (alphabetical for strings).
- The first parameter (`by`) expects the column(s) to sort by.
- If sorting by a single column, pass a string with its name.
- The `ascending` parameter customizes the sort order.
- The `na_position` parameter customizes where pandas places `NaN` values.

```python
# Creating a new nba dataframe:
newnba = pd.read_csv("nba.csv")
newnba.sort_values(by="Salary")
newnba.sort_values(by="Salary",na_position='first')
newnba.sort_values(by="Salary",na_position='first',ascending=False)
```

## Sort a DataFrame with the sort_values Method II
- To sort by multiple columns, pass the `by` parameter a list of column names. Pandas will sort in the specified column order (first to last).
- Pass the `ascending` parameter a Boolean to sort all columns in a consistent order (all ascending or all descending).
- Pass `ascending` a list to customize the sort order *per* column. The `ascending` list length must match the `by` list.

#### Passing multiple columns in `sort_values` is best illustrated when one or more columns have repeat values. As List being an ordered collection, the dataframe is sorted by the order of column names mentioned in the list.

```python
newnba.sort_values(by=["Team","Name"])
# Notice college column is sorted first & then the team column
newnba.sort_values(by=["College","Team"])
```

#### Each sorting within the list of sorting can also be controlled by passing a list of Booleans (true/false) into the `ascending` parameter

```python
newnba.sort_values(by=["College","Team"],ascending=[True,False])
# Notice the college column is sprted in ascending since its assigned parameter for ascending is True
# while the teams are sorted in descending since its assigned False for the ascending parameter
```

```python
# sorting positions in ascending and salaries by descending
newnba.sort_values(by=["Position","Salary"],ascending=[True,False],inplace=True)
```

## Rank Values with the rank Method
- The `rank` method assigns a numeric ranking to each **Series** value.
- Pandas will assign the same rank to equal values and create a "gap" in the dataset for the ranks.

```python
nba=pd.read_csv("nba.csv").dropna(how="all")
# Handling NaNs in Salary column 
nba["Salary"] = nba["Salary"].fillna(0).astype(int)
nba.info()
```

#### pandas `rank` func is similar to SQL -- both leave a gap in rank when there is a tie. If 2nd & 3d values are a tie, it'll be assgined ranks as 1,2,2,4 etc

```python
# Introducing a new column in the DataFrame
nba["SalaryRank"] = nba["Salary"].rank(ascending=False).astype(int)
```

```python
nba.sort_values(by=["Salary"],ascending=False)
```

```python
# Applying a dense rank to assign different ranks to ties
nba["SalaryRank1"]=nba["Salary"].rank(ascending=False,method='dense').astype(int)
nba.sort_values(by=["SalaryRank1"],ascending=False)
```
